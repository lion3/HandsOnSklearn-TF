```python
def NullValueProcess(DataFrame, KeyWords, MaxMissingSize=10):
    df = DataFrame.copy()
    k = MaxMissingSize
    m_indexs = list(df[df[KeyWords].values == 0].index)
    remain_indexs = []
    lost_index = []
    if len(m_indexs) != 0:
        cnt = 1
        init = 0
        for i in range(len(m_indexs)):
            df[df.values==0]=np.nan
            if i > 0:
                if m_indexs[i]==m_indexs[i-1]+1:
                    cnt += 1
                else:
                    if cnt >= k:
                        # tag missing data NA
                        for j in range(init,init+cnt+1):
                            lost_index.append(m_indexs[j])
                        cnt = 0
                        init = i
        remain_indexs = list(set(m_indexs)-set(lost_index))
        df = df.drop(lost_index) # drop 也是返回一个view
        df = df.fillna(method='ffill')
        #df = df.fillna(method='bfill')
    return df, lost_index, remain_indexs
```
